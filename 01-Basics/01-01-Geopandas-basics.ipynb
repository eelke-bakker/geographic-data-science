Using Geopandas, you can easily use ``gpd.read_file(path)`` to read all of these.

```python
# Let's read the countries file
countries = gpd.read_file('../Data/countries.gpkg')
countries.head()
```

```python
# Let's read the cities file
cities = gpd.read_file('../Data/cities.geojson')
cities.head()
```

It is often useful to directly plot the data on a map. Geopandas is built on top of Matplotlib and Descartes so maps can easily be made using a GeoDataFrame. The countries- and cities- dataframes above both contain the 'geometry' column which contains the geographical information. To plot, you can use ``GeoDataFrame.plot()``

```python
# Plot overlayed maps
fig, ax = plt.subplots(figsize=(12,12))
# This plots the countries
countries.plot(ax=ax)
# This adds the cities
cities.plot(markersize=0.5, color='red',ax=ax)
plt.tight_layout()
# You can uncomment the following line if you want to save your plot
#plt.savefig('overlay_map.png')
plt.show()
```

```python
# Let's read the rivers data
rivers = gpd.read_file('../Data/rivers.gpkg')

# Let's add the rivers to the previous plot
fig, ax = plt.subplots(figsize=(12,12))
# This plots the countries
countries.plot(ax=ax)
# This adds the cities
cities.plot(markersize=0.5, color='red',ax=ax)
rivers.plot(color='white',ax=ax)
plt.tight_layout()
# You can uncomment the following line if you want to save your plot
#plt.savefig('overlay_map.png')
plt.show()
```

## 2. Coordinate Reference Systems and Projections

The Coordinate Reference system (or CRS) tells us how our data as two dimensional (a plane) relates to actual places on earth (3 dimensional sphere). GeoDataFrames have a .crs attribute that you can check and adjust if necessary.

__NOTE__: It is very important, if you use multiple datasets, that all datasets use the same crs!

```python
# Let's check the crs of our three datasets
print(countries.crs)
print(cities.crs)
print(rivers.crs)

# Or use the following:
print('The countries dataset uses the same crs as the cities dataset: {}'.format((countries.crs == cities.crs)))
print('The countries dataset uses the same crs as the rivers dataset: {}'.format((countries.crs == rivers.crs)))
```

Different CRS and map projects use different measurements. For example, for some CRS, the coordinates are defined in decimal degrees (lat/lon), while others are defined in meters. You can always check [this website](https://map-projections.net/compare.php?p1=mercator-84&p2=robinson&sps=1) to compare and better understand map projections.

Let's try to reproject our data. We can do this using ``GeoDataFrame.to_crs({'init':'epsg:'})``

```python
# Currently, our data uses decimal degrees:
print(countries.geometry.head())

# If we want to project into the Mercator projection, we use epsg=3395
countries_projected = countries.to_crs({'init':'epsg:3395'})

# Check the new geometry column:
print(countries_projected.geometry.head())
```

Because we reprojected, the geometry is no longer in lat/lon style. Let's explore the difference by visualizing again. 

```python
# Show the  difference
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 16))
ax1 = countries.plot(ax=ax1)
ax1.set_title('EPSG=4326')
ax2 = countries_projected.plot(ax=ax2)
ax2.set_title('EPSG=3395')
plt.show()
```

So now what will happen if we overlay the cities again is that it will not properly allign. In the left plot, they cities do allign properly because both datasets have the same epsg, but in the right one, the cities have not yet been projected into the Mercator projection.

```python
# Show the  difference with the cities as well
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 16))
ax1 = countries.plot(ax=ax1)
cities.plot(markersize=0.5, color='red',ax=ax1)
ax1.set_title('EPSG=4326')
ax2 = countries_projected.plot(ax=ax2)
cities.plot(markersize=0.5, color='red',ax=ax2)
ax2.set_title('EPSG=3395')
plt.show()
```

```python
# So let's also reproject the cities:
cities_projected = cities.to_crs({'init':'epsg:3395'})
# Show the  difference with the cities as well
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(20, 16))
ax1 = countries.plot(ax=ax1)
cities.plot(markersize=0.5, color='red',ax=ax1)
ax1.set_title('EPSG=4326')
ax2 = countries_projected.plot(ax=ax2)
cities_projected.plot(markersize=0.5, color='red',ax=ax2)
ax2.set_title('EPSG=3395')
plt.show()

# Okay it looks a bit funny bleghh
```
